```python
import matplotlib.pyplot as plt
```

To orient a bar chart in Matplotlib, you can control the orientation of the bars by specifying the width parameter for vertical bars or the height parameter for horizontal bars in the bar() function.

```python
plt.bar(["Europe","America","Japan","Australia","Newyork"],[10,20,30,40,50])
plt.title(label="Demo Bar Chart")
plt.xlabel(xlabel="X-Axis")
plt.ylabel(ylabel="Y-Axis")
plt.show()
```

```python
plt.barh(["Europe","America","Japan","Australia","Newyork"],[10,20,30,40,50],height=0.5)
plt.title(label="Demo Bar Chart")
plt.xlabel(xlabel="X-Axis")
plt.ylabel(ylabel="Y-Axis")
plt.show()
```

```python

plt.figure(figsize=(8, 8)) 
plt.pie([10,20,30,40,50],labels=["Europe","America","Japan","Australia","Newyork"],autopct='%1.1f%%', startangle=140)
plt.title(label="Demo Pie Chart")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
sizes = [30, 20, 25, 15, 10]  # Sizes of different sections of the pie chart
labels = ['A', 'B', 'C', 'D', 'E']  # Labels for each section

plt.figure(figsize=(8, 8))  # Optional: Set the figure size

# Plotting the pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title('Sample Pie Chart')
plt.show()
```

```python
plt.figure(figsize=(8, 8)) 
plt.plot(sizes, labels, marker='o', linestyle='-', color='b', label='Line')
plt.title(label="Demo Pie Chart")
plt.show()
```
